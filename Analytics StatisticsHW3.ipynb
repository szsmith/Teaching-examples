```python
# Shapiro-Wilk Test
from numpy.random import seed
from numpy.random import randn
from scipy.stats import shapiro
# seed the random number generator
seed(1)
# normality test
stat, p = shapiro(df['DAM'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

```python
z_critical = 1.96 # alpha level of 0.05 and two-tailed test
x_bar = df['DAM'].mean()
N = 100
SE = sigma/np.sqrt(N)
z_stat = (x_bar - mu)/SE
print(z_stat)
```

```python
# since the z_stat is less than Z_critical we accept the null hypothesis
# and reject the alternative. Statistically, we say the sample mean is no 
# different than the population mean and thus the sample is drawn from 
# the population.
```
