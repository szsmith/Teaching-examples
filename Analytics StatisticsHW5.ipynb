```python
import pandas as pd

df1 = pd.read_csv("hw5_data.csv")
df1.head(5)

# Recoding value from numeric to string
df1['group'].replace({1: 'psy', 2: 'stats', 3: 'art'}, inplace= True)
df1.head(5)
```

```python

# Note the difference in argument order
model = ols('activities ~ C(group)', data=df1)

results=model.fit()
#predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
results.summary()

# See the results differnce now with a base category
```

```python
import statsmodels.api as sm
aov_table = sm.stats.anova_lm(results, typ=2)
aov_table
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(df1['activities'], df1['group'])
mc_results = mc.tukeyhsd()
print(mc_results)
#tukeyhsd include the mean difference and variance intervals for the studentized range statistic
```
